```python
def nn_pscore_model1(input_size = 36):
    '''
    Returns a compiled keras model
    '''
    model = Sequential()
    model.add(Dense(30, input_dim = input_size, activation = 'relu'))
    model.add(Dense(30, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # model.name="NAME"
    return model
```

```python
def nn_pscore_model2(input_size = 36):
    '''
    Returns a compiled keras model
    '''
    model = Sequential()
    model.add(Dense(40, input_dim = input_size, activation = 'relu'))
    model.add(Dense(20, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # model.name="NAME"
    return model
```

```python
def nn_pscore_model3(input_size = 36):
    '''
    Returns a compiled keras model
    '''
    model = Sequential()
    model.add(Dense(40, input_dim = input_size, activation = 'relu'))
    model.add(Dense(20, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # model.name="NAME"
    return model
```

```python
def nn_pscore_model4(input_size = 36):
    '''
    Returns a compiled keras model
    
    This model implements dropouts
    '''
    model = Sequential()
    model.add(Dense(30, input_dim = input_size, activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(30, activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # model.name="NAME"
    return model
```

```python
def nn_pscore_model5(input_size = 36):
    '''
    Returns a compiled keras model
    
    This model implements dropouts
    '''
    model = Sequential()
    model.add(Dense(40, input_dim = input_size, activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(30, activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(20, activation = 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation = 'sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # model.name="NAME"
    return model
```
